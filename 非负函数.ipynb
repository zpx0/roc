```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    fx = 2 * x
    fy = 2 * y
    return np.array([fx, fy])

learning_rate = 0.1
num_iterations = 20

x0 = 1.0
y0 = 1.0


loss_values = []
x_values = []
y_values = []

for i in range(num_iterations):
    grad = grad_f(x0, y0)
    loss = f(x0, y0)
    loss_values.append(loss)
    x_values.append(x0)
    y_values.append(y0)
    

    x0 -= learning_rate * grad[0]
    y0 -= learning_rate * grad[1]
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(loss_values, label='Loss')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
```

```python
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)
ax.plot(x_values, y_values, [f(xi, yi) for xi, yi in zip(x_values, y_values)], color='r', label='Gradient Descent Path')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gradient Descent Path on f(x, y) = x**2+ y**2')
ax.legend()
plt.show()
```
