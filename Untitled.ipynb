```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib import rcParams
```

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
x1 = torch.tensor(0.0, requires_grad=True)
x2 = torch.tensor(1.0, requires_grad=True)
```

```python
v0=x1
v1=x2
v2= torch.tensor(1., requires_grad=True)
v3=v0**2+2*v1+v2
v4=sigmoid(v3)
v5=4*v4+v2
y=sigmoid(v5)
```

```python
v1 = x1**2 + 2*x2 + 1
s1 = sigmoid(v1)
s2 = 3*s1
s3 = s1
v3 = s2 + s3 + 1
y = sigmoid(v3)
```

```python
node_dict = {"v0": v0,"v1": v1, "v2": v2,"v3": v3,"v4": v4, "v5": v5,  }
var_dict = {"x1": x1, "x2": x2}
```

```python
for node_name in node_dict:
    for var_name in var_dict:
        node = node_dict[node_name]
        var = var_dict[var_name]
        
        if var.grad is not None:
            var.grad.zero_()
        else:
            var.grad = torch.tensor(0.0)
        node.backward(retain_graph=True)
        print(f"节点{node_name}对{var_name}导数: ", var.grad)
```

```python
G = nx.DiGraph()

G.add_node("x1", label="v0")
G.add_node("x2", label="v1")
G.add_node("1", label="v2")
G.add_node("v0**2+2*v1+v2", label="v3")
G.add_node("sigmoid(v3)", label="v4")
G.add_node("4*v4+v2", label="v5")
G.add_node("sigmoid(v5)", label="y")

G.add_edges_from([
     ("x1", "v0**2+2*v1+v2"),  
    ("x2", "v0**2+2*v1+v2"), 
    ("1", "v0**2+2*v1+v2"),
    ("v0**2+2*v1+v2", "sigmoid(v3)"),
    ("sigmoid(v3)", "4*v4+v2"),  
    ("4*v4+v2", "sigmoid(v5)")   
])

rcParams['font.sans-serif'] = ['SimHei'] 
rcParams['axes.unicode_minus'] = False 

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=8, font_weight="bold", arrowsize=20)
plt.title("链式图")
plt.show()
```
